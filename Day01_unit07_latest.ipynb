## 우리동네 인구 구조 시각화하기

```python
import csv
import matplotlib.pyplot as plt

class spread_of_pops:
    def __init__(self, csv_path):
        self.csv_path = csv_path

    def read_local_data(self, local_name):
        f = open('age.csv')
        data = csv.reader(f)

        # 검색되는 지역명의 Row 가 여러개인 경우 잘못된 자료가 나오는 것을 방지하기 위해 
        # 0~100까지 나이별 Array 를 만들고 해당 방에 값을 계속 누적하여 보여준다.
        self.spread_ages = []
        for i in range(101):
            self.spread_ages.append(0)

        for row in data:
            if ((local_name in row[0]) and (row[0][-6:] != '00000)')): # 동이 아닌 상위 지역명을 입력시 상위지역 합계는 제외한다.
                idx = 0
                for i in row[3:]:
                    n = i.replace(',','')
                    self.spread_ages[idx] += int(n)
                    idx += 1
        
        f.close()
    
    def show(self, style = 'ggplot'):
        plt.style.use(style)
        plt.plot(self.spread_ages)
        plt.show()
        
    def row(self):
        print(self.spread_ages)
```

```python
sp = spread_of_pops('age.csv')
```

```python
sp.read_local_data('신도림')
sp.row()
sp.show()
```
