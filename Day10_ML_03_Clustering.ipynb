```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
```

```python
X, y = make_blobs(random_state=1)
X.shape, y.shape
```

```python
agg = AgglomerativeClustering(n_clusters=3)
p = agg.fit_predict(X)
```

```python
plt.rc('font', family='Malgun Gothic')
mglearn.discrete_scatter(X[:, 0], X[:, 1], p)
plt.legend(['클러스터0', '클러스터1', '클러스터2'], loc='best')
plt.xlabel('특성 0')
plt.ylabel('특성 1')
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters=3).fit(X)
p = km.predict(X)
```

```python
plt.rc('font', family='Malgun Gothic')
mglearn.discrete_scatter(X[:, 0], X[:, 1], p)
plt.legend(['클러스터0', '클러스터1', '클러스터2'], loc='best')
plt.xlabel('특성 0')
plt.ylabel('특성 1')
plt.show()
```

### DBSCAN

```python
mglearn.plots.plot_dbscan()
```

```python
from sklearn.cluster import DBSCAN
```

```python
p = DBSCAN().fit_predict(X)
```

```python
mglearn.discrete_scatter(X[:, 0], X[:, 1], p)
plt.show()
```

```python
# eps:샘플들간의 거리, min_samples:클러스터가 되기 위한 최소 갯수 > eps내의 min_samples개를 하나의 클러스터로 지정
p = DBSCAN(eps=1.5, min_samples=5).fit_predict(X)
mglearn.discrete_scatter(X[:, 0], X[:, 1], p)
plt.show()
```

```python
p = DBSCAN(eps=1.0, min_samples=5).fit_predict(X)
mglearn.discrete_scatter(X[:, 0], X[:, 1], p)
plt.show()
```
